```python
from src.algorithms import NeedlemanWunsch, SmithWaterman
```

```python
import numpy as np

# Sample 10 tuples of sequences from letters A, T, G, C
for i in range(10):
    np.random.seed(i)
    len_seq1 = np.random.randint(3, 10)
    len_seq2 = np.random.randint(3, 10)
    seq1 = ''.join(np.random.choice(['A', 'T', 'G', 'C'], len_seq1))
    seq2 = ''.join(np.random.choice(['A', 'T', 'G', 'C'], len_seq2))
    print(f"Sequence 1: {seq1}")
    print(f"Sequence 2: {seq2}")
    nw = NeedlemanWunsch(3, seq1, seq2, 'substitution_matrix.csv', f'./results/{seq1}_{seq2}.txt')
    nw.init_matrix()
    nw.fill_matrix()
    nw.print_matrix()
    nw.traceback(save=True, save_tracebacks=True)

    sw = SmithWaterman(3, seq1, seq2, 'substitution_matrix.csv', f'./results/{seq1}_{seq2}.txt')
    sw.init_matrix()
    sw.fill_matrix()
    sw.print_matrix()
    sw.traceback(save=True, save_tracebacks=True)
```
